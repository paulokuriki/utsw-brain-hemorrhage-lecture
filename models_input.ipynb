### 1. Finding the Perfect Sweetness Level to Distinguish Apples from Oranges

Imagine you're blindfolded and given two types of fruits to taste – apples and oranges. You realize that apples tend to be a bit sweeter than oranges. So, you come up with a game plan. Based on the sweetness level, you'll guess which fruit you're tasting.

### The Game Plan
1. Gather Data: You've tasted a few apples and oranges and recorded their sweetness levels. For apples, you recorded sweetness levels like `6, 7, 8, 7.5, 6.5`, and for oranges, you got values like `3, 4, 5, 4.5, 3.5, 6, 7.`

2. The Sweetness Threshold: This is the level of sweetness you'll use as a reference. If a fruit's sweetness is above this level, you'll guess it's an apple. If it's below, you'll guess it's an orange.

3. Goal is Finding the Best Threshold: You want to be as accurate as possible in your guesses. So, you'll try different sweetness levels as your threshold and see which one gives you the most correct guesses.

```python
# Training data
apples_sweetness = [6, 7, 8, 7.5, 6.5]
oranges_sweetness = [3, 4, 5, 4.5, 3.5, 6, 7]

# Classifier function
def classify_fruit(sweetness, threshold):
    if sweetness > threshold:
        return "apple"
    else:
        return "orange"

# Determine the best threshold
best_threshold = None
best_accuracy = 0

for potential_threshold in [i * 0.1 for i in range(10, 100)]:    
    correct_guesses = 0
    
    for sweetness in apples_sweetness:
        if classify_fruit(sweetness, potential_threshold) == "apple":
            correct_guesses += 1

    for sweetness in oranges_sweetness:
        if classify_fruit(sweetness, potential_threshold) == "orange":
            correct_guesses += 1

    accuracy = correct_guesses / len(apples_sweetness + oranges_sweetness)  # total_guesses
    print(f'For a threashold of {potential_threshold:.1f}, the accuracy was: {int(accuracy*100)}%')
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_threshold = potential_threshold
        print(f'New best threashold is {potential_threshold:.1f}, with an accuracy was: {int(accuracy*100)}%')

print(f"Best threshold for classification is: {best_threshold}")
```

## Testing Our Sweetness Guessing Game

After setting up our game plan in the first part, now it's time to see how good we are at our fruit guessing game!

### Gathering Fresh Fruits for Testing
We've grabbed some new apples and oranges that we haven't tasted before. 

For **apples**, we have sweetness values like `4.5, 6.2, 7.1, 7.8, 7.3`, and for **oranges**, we have `3.2, 4.1, 4.8, 3.7, 6.1.`

### Let's Start Guessing!
Using our best sweetness threshold (the one we found earlier), we'll guess each fruit's identity. Is it sweeter than our threshold? We guess it's an apple. Less sweet? We guess it's an orange.

### How Well Did We Do?
To understand how accurate our guesses were, we'll use a few measures:

- Accuracy: How many fruits did we guess correctly?
- Sensitivity (Recall): Of all the apples, how many did we correctly guess as apples?
- Specificity: Of all the oranges, how many did we correctly guess as oranges?
- Predictive Positive (Precision): When we guessed a fruit is an apple, how often were we right?
- Predictive Negative: When we guessed a fruit is an orange, how often were we right?

We'll print these measures to see our performance.

### A Picture Speaks a Thousand Words
To make it even easier to understand, we'll visualize our results using a colorful chart called a Confusion Matrix. This chart will show:

- How many apples we correctly guessed as apples.
- How many oranges we mistakenly guessed as apples.
-How many apples we mistakenly guessed as oranges.
- How many oranges we correctly guessed as oranges.

The bigger the diagonal values in the chart, the better we are at our game!

### Let's See the Results!
Now, we'll display this colorful chart and see how well we did in our guessing game. Are you ready?

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
import itertools

# New Test data
test_apples_sweetness = [4.5, 6.2, 7.1, 7.8, 7.3]
test_oranges_sweetness = [3.2, 4.1, 4.8, 3.7, 6.1]

# Predictions based on best threshold
y_true = [1 if label == 'apple' else 0 for label in ["apple"] * len(test_apples_sweetness) + ["orange"] * len(test_oranges_sweetness)]
y_pred = [1 if classify_fruit(sweetness, best_threshold) == "apple" else 0 for sweetness in test_apples_sweetness + test_oranges_sweetness]

# Compute metrics
conf_mat = confusion_matrix(y_true, y_pred)
TN, FP, FN, TP = conf_mat.ravel()

accuracy = accuracy_score(y_true, y_pred)
sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
predictive_pos = TP / (TP + FP)
predictive_neg = TN / (TN + FN)

print(f"Accuracy: {accuracy:.2f}")
print(f"Sensitivity (Recall): {sensitivity:.2f}")
print(f"Specificity: {specificity:.2f}")
print(f"Predictive Positive (Precision): {predictive_pos:.2f}")
print(f"Predictive Negative: {predictive_neg:.2f}")

# Plot confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

# Display the matrix
plot_confusion_matrix(conf_mat, classes=['Orange', 'Apple'], title='Confusion matrix')
```

## 2. The Advanced Fruit Guessing Game: Using 3 Clues!

Previously, we played a game where we guessed if a fruit was an apple or a banana based on its sweetness. This time, our game is a tad more advanced. Instead of just sweetness, we're using three different clues about each fruit to make our guess.

Imagine each fruit tells us three secrets about themselves:

1. How `sweet` they are.
2. How `colorful` they look.
3. How `firm` they feel when touched.

### Our Pile of Fruits

We've got a fresh pile of apples and bananas, and they've whispered their secrets to us:

- Apples have given clues like: `8 sweet, 8 colorful, 5 firm`.
- Bananas have clues like: `2 sweet, 12 colorful, 9 firm`.

### The Game Plan

To guess the fruits:

1. We'll give some importance or weight to each of the three secrets.
2. We'll calculate a total score based on the clues and the importance we've assigned to them.
3. If the score is above a special number (called the threshold), we'll guess it's an apple. Otherwise, it's a banana.

### Training Time: Adjusting Our Strategy

Just like a sportsperson trains to improve, we'll train our guessing strategy to become better.

- Initially, our weights or importance levels for the secrets are all set to zero.
- We then test our current strategy on our pile of fruits.
- Based on how well (or poorly) we guess, we adjust our weights and the special threshold.
- We repeat this several times, each time fine-tuning our strategy.
This way, we hope that by the end of our training, we have a really good strategy to tell apart apples from bananas!

### The Big Reveal
After we finish training, we'll reveal the importance we've given to each secret (the weights) and our special threshold. With these, we are all set to play our advanced fruit guessing game!

```python
from tqdm import tqdm

# Dataset
apples_data = [[8, 8, 5], [7, 7, 6], [9, 7.5, 5.5], [8.5, 8, 6], [7.5, 8.5, 5]]
bananas_data = [[2, 12, 9], [3, 11.5, 9.5], [3.5, 12.5, 8.5], [2.5, 11, 9], [3, 13, 10]]

data = apples_data + bananas_data
labels = [1] * len(apples_data) + [0] * len(bananas_data)  # 1 for apples, 0 for bananas

# Simple linear classifier
def classify(fruit, weights, threshold):
    score = sum([f*w for f, w in zip(fruit, weights)])
    return 1 if score > threshold else 0  # 1 for apple, 0 for banana

# Training process
weights = [0, 0, 0]
threshold = 0
learning_rate = 0.1
iterations = 1000

for _ in range(iterations):
    for fruit, label in zip(data, labels):
        prediction = classify(fruit, weights, threshold)
        error = label - prediction
        
        # Adjust weights and threshold based on error
        for i in range(3):  # we have 3 features
            weights[i] += learning_rate * error * fruit[i]
        threshold -= learning_rate * error

print(f"Trained weights: {weights}")
print(f"Threshold: {threshold}")
```

After our intense training session, we're now ready to test our improved fruit-guessing skills! We've gathered a new set of apples and bananas, and we're keen to see how well our strategy works on them.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
import itertools

# New Test data
test_apples_data = [[5, 7, 6], [7.5, 7.2, 6], [8.5, 8, 6], [9, 8.5, 5.5], [7, 8.5, 5.5]]
test_bananas_data = [[3, 11, 9], [2.5, 12, 8.5], [3.5, 12.5, 9], [2, 11.5, 10], [3.5, 12, 9.5]]

# Predictions based on best weights and threshold
y_true = [1 if label == 'apple' else 0 for label in ["apple"] * len(test_apples_data) + ["banana"] * len(test_bananas_data)]
y_pred = [1 if classify(fruit, weights, threshold) == 1 else 0 for fruit in test_apples_data + test_bananas_data]

# Compute metrics
conf_mat = confusion_matrix(y_true, y_pred)
TN, FP, FN, TP = conf_mat.ravel()

accuracy = accuracy_score(y_true, y_pred)
sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
predictive_pos = TP / (TP + FP)
predictive_neg = TN / (TN + FN)

print(f"Accuracy: {accuracy:.2f}")
print(f"Sensitivity (Recall): {sensitivity:.2f}")
print(f"Specificity: {specificity:.2f}")
print(f"Predictive Positive (Precision): {predictive_pos:.2f}")
print(f"Predictive Negative: {predictive_neg:.2f}")

# Plot confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

# Display the matrix
plot_confusion_matrix(conf_mat, classes=['Banana', 'Apple'], title='Confusion matrix')
```

## 3. Understanding Image Data for Machine Learning
### What is an Image in the context of Machine Learning?
In the realm of Machine Learning and particularly Computer Vision, an image is often treated as a matrix of pixel values. These pixel values range from 0 to 255 in the case of 8-bit grayscale images. The value indicates the intensity of the pixel, with 0 being black and 255 being white.

### Transforming Image Data for Model Input
When using images for machine learning models, particularly those models that don't inherently deal with 2D data (like many classical ML algorithms), these 2D pixel matrices need to be converted or flattened into a 1D array. Each pixel of the image becomes a unique feature or input for the model.

For instance, consider an image that is resized to 20x20 pixels. This image has a total of 400 pixels. When flattened, it becomes an array (or a row in a dataset) with 400 feature columns, each column representing a unique pixel from the image.

## Resizing Images for Machine Learning
In machine learning, particularly when dealing with images, it's imperative that all input data have a consistent structure. This means that all images used to train, validate, or test a model must have the same dimensions. One common way to ensure this is by resizing all images to a specific width and height.

### Why Resize?

- Consistent Input Size: Neural networks and many machine learning algorithms require fixed-size input. This means that all images fed into a model must have the same number of features, which corresponds to the number of pixels in this context.
- Computational Efficiency: Processing large images can be computationally expensive. Resizing images to a smaller size can make computations more manageable and training faster, without significantly compromising the quality of the training process.
- Memory Efficiency: Storing large images requires more memory. Especially when working with large datasets, this can become problematic. Resized images consume less memory, making it feasible to work with large image datasets on standard hardware.

```python
from PIL import Image
import matplotlib.pyplot as plt

def display_resized_image(filename, size=(20, 20)):
    # Open the image
    img = Image.open(filename)
    
    # Convert to grayscale
    img_gray = img.convert('L')
    
    # Resize it
    img_resized = img_gray.resize(size)
    
    # Display the image
    plt.imshow(img_resized, cmap='gray')  # Specify colormap as 'gray'
    plt.axis('off')  # Hide axis
    plt.show()

# Now, display both images
display_resized_image('hemorrhage.jpeg')
display_resized_image('normal.jpeg')
```

## Pixels as Features in Image Processing
When we think of images in the context of machine learning, we often view them as 2D arrays (or matrices) of numbers. Each number in this matrix corresponds to the intensity of a pixel in the image. This pixel intensity can be a value between 0 and 255, with 0 representing black and 255 representing white (for grayscale images).

### Why Convert Images to Pixel Values?

- Quantifiable Representation: Machine learning models require numerical input. By representing an image as a matrix of pixel intensities, we convert visual information into a format that algorithms can understand and process.
- Feature Extraction: Each pixel in this matrix becomes a feature. For a 20x20 image, we have 400 features that the algorithm will use to make predictions or classifications.

```python
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt

def display_image_pixels(filename, size=(20, 20)):
    # Open the image
    img = Image.open(filename)
    
    # Convert to grayscale
    img_gray = img.convert('L')
    
    # Resize it
    img_resized = img_gray.resize(size)
    
    # Convert to numpy array
    pixel_data = np.array(img_resized)
    
    # Convert numpy array to pandas DataFrame
    df = pd.DataFrame(pixel_data)
    
    # Display the DataFrame
    display(df)

# Display pixels of both images in tables
display_image_pixels('hemorrhage.jpeg')
display_image_pixels('normal.jpeg')
```

```python
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt

def display_image_pixels(filename, size=(20, 20)):
    # Open the image
    img = Image.open(filename)
    
    # Convert to grayscale
    img_gray = img.convert('L')
    
    # Resize it
    img_resized = img_gray.resize(size)
    
    # Convert to numpy array
    pixel_data = np.array(img_resized)
    
    # Convert numpy array to pandas DataFrame
    df = pd.DataFrame(pixel_data)
    
    # Function to color the cell background based on its value
    def color_cells(val):
        color = f'rgb({val}, {val}, {val})'  # Grayscale is where R = G = B
        return f'background-color: {color};'
    
    # Apply the styling and display
    styled = df.style.applymap(color_cells)
    display(styled)

# Display pixels of both images with cell background color
display_image_pixels('hemorrhage.jpeg')
display_image_pixels('normal.jpeg')
```

```python
from PIL import Image
import numpy as np

def image_to_flattened_array(filename, size=(20, 20)):
    """Convert an image to a flattened numpy array of grayscale pixel values."""
    img = Image.open(filename)
    img_gray = img.convert('L')
    img_resized = img_gray.resize(size)
    pixel_data = np.array(img_resized)
    return pixel_data.ravel()

# Convert the images to flattened numpy arrays
hemorrhage_array = image_to_flattened_array('hemorrhage.jpeg')
normal_array = image_to_flattened_array('normal.jpeg')

# Print the shapes
print("Shape of hemorrhage image:", hemorrhage_array.shape)
print("Shape of normal image:", normal_array.shape)
print("\nhemorrhage_array:")
print(hemorrhage_array)
print("\nnormal_array:")
print(normal_array)
```
